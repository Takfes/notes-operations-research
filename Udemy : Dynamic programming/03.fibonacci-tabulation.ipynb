```python
import time
```

```python
def fib(n, verbose=False, table=None):
    if table is None:
        table = [None] * (n + 1)
    table[0] = 0
    table[1] = 1
    for i in range(2, n + 1):
        table[i] = table[i - 1] + table[i - 2]
        if verbose:
            print(table)
    return table[n]
```

```python
results_without_tabulation = []
for i in range(20):
    start = time.perf_counter()
    fib(i)
    end = time.perf_counter()
    results_without_tabulation.append((i,end - start))
```
